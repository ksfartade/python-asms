```python
# Question 1: What is Simple Linear Regression?
"""
Simple Linear Regression is a statistical method used to model the linear relationship between a dependent variable (often denoted as Y) and a single independent variable (often denoted as X).

The goal of Simple Linear Regression is to find the best-fitting straight line that describes how the dependent variable changes as the independent variable changes. This line is represented by the equation:

Y = β₀ + β₁X + ε

Where:
- Y is the dependent variable
- X is the independent variable
- β₀ is the y-intercept (the value of Y when X is 0)
- β₁ is the slope of the line (the change in Y for a one-unit change in X)
- ε is the error term (the difference between the observed value of Y and the value predicted by the line)

The method aims to estimate the values of β₀ and β₁ that minimize the sum of the squared differences between the observed Y values and the values predicted by the line. This minimization process is typically done using the method of Ordinary Least Squares (OLS).
"""
```

```python
# Question 2: What are the key assumptions of Simple Linear Regression?
"""
The key assumptions of Simple Linear Regression are:

1. Linearity: The relationship between the independent variable (X) and the dependent variable (Y) is linear. This means that the relationship can be represented by a straight line.

2. Independence: The observations are independent of each other. There should be no relationship between the residuals (the differences between the observed and predicted values) for different observations.

Violations of these assumptions can affect the validity and reliability of the regression results. It's important to check these assumptions when performing linear regression.
"""
```

```python
# Question 3: What is heteroscedasticity, and why is it important to address in regression models?
"""
Heteroscedasticity is the violation of the homoscedasticity assumption in regression analysis. It occurs when the variance of the error term (residuals) is not constant across all levels of the independent variable(s). In other words, the spread of the residuals is not the same for all values of X.

Visually, in a scatter plot of residuals versus predicted values or independent variables, heteroscedasticity might appear as a fanning-out or fanning-in pattern of the residuals.

Why is it important to address in regression models?

1. Inefficient OLS Estimators: When heteroscedasticity is present, the Ordinary Least Squares (OLS) estimators of the regression coefficients are still unbiased and consistent, but they are no longer the most efficient estimators. This means that the standard errors of the coefficients will be biased, leading to incorrect inferences about the statistical significance of the predictors.

2. Invalid Hypothesis Tests and Confidence Intervals: Because the standard errors are biased, hypothesis tests (like t-tests and F-tests) and confidence intervals will be unreliable. You might incorrectly conclude that a predictor is statistically significant when it is not, or vice versa. This can lead to wrong conclusions about the relationships between variables.

3. Misleading Model Fit: Heteroscedasticity can affect measures of model fit, such as R-squared. While R-squared itself is not directly biased by heteroscedasticity, the inferences drawn from it regarding the proportion of variance explained can be misleading due to the unreliable standard errors.

Addressing heteroscedasticity is important to ensure the validity of statistical inferences and the reliability of the regression model's results. Methods to address heteroscedasticity include:

- Transforming the dependent variable (e.g., using a logarithmic transformation).
- Using weighted least squares (WLS) regression, which gives less weight to observations with larger variances.
- Using robust standard errors (also known as heteroscedasticity-consistent standard errors), which provide valid standard errors even in the presence of heteroscedasticity.
"""
```

```python
# Question 4: What is Multiple Linear Regression?
"""
Multiple Linear Regression is an extension of Simple Linear Regression that models the linear relationship between a dependent variable (Y) and two or more independent variables (X₁, X₂, ..., Xn).

The goal is to find the best-fitting linear equation that describes how the dependent variable changes as the independent variables change simultaneously. The equation for Multiple Linear Regression is:

Y = β₀ + β₁X₁ + β₂X₂ + ... + βnXn + ε

Where:
- Y is the dependent variable
- X₁, X₂, ..., Xn are the independent variables
- β₀ is the y-intercept (the value of Y when all independent variables are 0)
- β₁, β₂, ..., βn are the coefficients for each independent variable (the change in Y for a one-unit change in the corresponding independent variable, holding all other independent variables constant)
- ε is the error term

Similar to Simple Linear Regression, the method aims to estimate the values of the coefficients (β₀, β₁, ..., βn) that minimize the sum of the squared differences between the observed Y values and the values predicted by the linear equation. This is also typically done using the method of Ordinary Least Squares (OLS).

Multiple Linear Regression is used for:
- Understanding the combined effect of multiple independent variables on a dependent variable
- Predicting the value of the dependent variable based on multiple independent variables
- Identifying which independent variables have a significant linear relationship with the dependent variable, while controlling for the effects of other predictors

Key assumptions of Multiple Linear Regression include those of Simple Linear Regression (linearity, independence, homoscedasticity, normality of residuals) plus:
- No perfect multicollinearity: The independent variables should not be perfectly linearly related to each other. High multicollinearity can make it difficult to estimate the individual coefficients accurately.
"""
```

```python
# Question 5: What is polynomial regression, and how does it differ from linear regression?
"""
Polynomial Regression is a form of regression analysis in which the relationship between the independent variable (X) and the dependent variable (Y) is modeled as an nth degree polynomial. While it is still considered a form of linear regression because the model is linear in the coefficients (β), it allows for fitting non-linear relationships between the variables.

The equation for a simple polynomial regression with one independent variable is:

Y = β₀ + β₁X + β₂X² + ... + βnXⁿ + ε

Where:
- Y is the dependent variable
- X is the independent variable
- β₀, β₁, ..., βn are the coefficients
- n is the degree of the polynomial
- ε is the error term

How does it differ from linear regression?

The key difference lies in the form of the relationship being modeled:

1. Linear Regression: Models a strictly linear relationship between the independent and dependent variables. The equation is a straight line (or a hyperplane in multiple linear regression).

2. Polynomial Regression: Models a non-linear relationship between the independent and dependent variables by including polynomial terms (X², X³, etc.) in the equation. This allows the model to capture curves and bends in the data that a simple linear model cannot.

Despite modeling a non-linear relationship between X and Y, polynomial regression is considered "linear" in the sense that the model is linear in the parameters (β coefficients). This means that the estimation methods used for linear regression (like Ordinary Least Squares) can still be applied to estimate the coefficients in a polynomial regression model.

Polynomial regression is useful when the relationship between variables is not strictly linear and exhibits curvature. However, using higher-degree polynomials can lead to overfitting the data, so care must be taken in choosing the appropriate degree of the polynomial.
"""
```

```python
# Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
# Plot the regression line over the data points.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for scikit-learn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X)

# Get the coefficients
beta_0 = model.intercept_
beta_1 = model.coef_[0]

print(f"Intercept (β₀): {beta_0}")
print(f"Coefficient (β₁): {beta_1}")

# Plot the data points and the regression line
plt.scatter(X, Y, label='Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 7: Fit a Multiple Linear Regression model on this sample data:
# ● Area = [1200, 1500, 1800, 2000]
# ● Rooms = [2, 3, 3, 4]
# ● Price = [250000, 300000, 320000, 370000]
# Check for multicollinearity using VIF and report the results.
# (Include your Python code and output in the code box below.)

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm # Import statsmodels.api with alias sm

# Sample data
data = {'Area': [1200, 1500, 1800, 2000],
        'Rooms': [2, 3, 3, 4],
        'Price': [250000, 300000, 320000, 370000]}

df = pd.DataFrame(data)

# Define independent variables (X) and dependent variable (Y)
X = df[['Area', 'Rooms']]
Y = df['Price']

# Create a Multiple Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Print the coefficients
print("Multiple Linear Regression Model:")
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Check for multicollinearity using VIF
# Add a constant to the independent variables for VIF calculation
X_with_constant = sm.add_constant(X)

# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data["feature"] = X_with_constant.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_constant.values, i)
                  for i in range(X_with_constant.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

```python
# Question 8: Implement polynomial regression on the following data:
# ● X = [1, 2, 3, 4, 5]
# ● Y = [2.2, 4.8, 7.5, 11.2, 14.7]
# Fit a 2nd-degree polynomial and plot the resulting curve.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for scikit-learn
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (2nd degree)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the polynomial features
model.fit(X_poly, Y)

# Predict Y values using the fitted model
Y_pred = model.predict(X_poly)

# Plot the data points and the polynomial regression curve
plt.scatter(X, Y, label='Data Points')
plt.plot(X, Y_pred, color='red', label='Polynomial Regression Curve (Degree 2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 9: Create a residuals plot for a regression model trained on this data:
# ● X = [10, 20, 30, 40, 50]
# ● Y = [15, 35, 40, 50, 65]
# Assess heteroscedasticity by examining the spread of residuals.
# (Include your Python code and output in the code box below.)

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Create residuals plot
plt.scatter(Y_pred, residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()

# Assess heteroscedasticity
print("Assessing Heteroscedasticity:")
print("Examine the residuals plot. If the spread of residuals is roughly constant across all predicted values, then the assumption of homoscedasticity is met. If the spread of residuals changes (e.g., fans out or in), then heteroscedasticity is present.")
```

*   **Check VIF:** Calculate Variance Inflation Factors (VIF) to quantify correlation.
*   **Remove or Combine:** Remove highly correlated variables or combine them.
*   **Regularization:** Use techniques like Ridge or Lasso regression.

*   **Validate Assumptions:** Re-check regression assumptions.
*   **Handle Outliers:** Identify and address influential data points.
